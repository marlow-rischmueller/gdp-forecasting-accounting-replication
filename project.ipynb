```python
# Compute Variants for feature set 2
# DataFrame für die neuen Werte
df_feature_set_2_calculated = pd.DataFrame()

# Berechnung der spezifizierten Wachstumsmaße für jede Variable
for variable, calculations in tqdm(berechnungen_feature_set2.items()):
    if 'sd_dif' in calculations:
        # Standardabweichung der Quarter-over-Quarter growth rate ("sd_dif")
        if f'{variable.lower()}_dif' not in feature_data_raw.columns:
           feature_data_raw[f'{variable.lower()}_dif'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff()
        df_feature_set_2_calculated[f'{variable.lower()}_sd_dif'] = feature_data_raw.groupby(['fyearq', 'fqtr'])[f'{variable.lower()}_dif'].std()

    if 'sd_yoy' in calculations:
        # Standardabweichung der Year-over-Year growth rate ("sd_yoy")
        if f'{variable.lower()}_yoy' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_yoy'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff(4)
        df_feature_set_2_calculated[f'{variable.lower()}_sd_yoy'] = feature_data_raw.groupby(['fyearq', 'fqtr'])[f'{variable.lower()}_yoy'].std()

    if 'vwmean_dif' in calculations:
        # Value-weighted average der Quarter-over-Quarter growth rate ("vwmean_dif")
        if f'{variable.lower()}_dif' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_dif'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff()
        feature_data_raw[f'{variable.lower()}_dif_mkvaltq'] = (feature_data_raw[f'{variable.lower()}_dif'] * feature_data_raw['mkvaltq'])
        df_feature_set_2_calculated[f'{variable.lower()}_vwmean_dif'] = feature_data_raw.groupby(['fyearq', 'fqtr'])[f'{variable.lower()}_dif_mkvaltq'].sum()/feature_data_raw.groupby(['fyearq', 'fqtr'])['mkvaltq'].sum()

    if 'mean_dif' in calculations:
        # Cross-sectional equal-weighted average der Quarter-over-Quarter growth rate ("mean_dif")
        if f'{variable.lower()}_dif' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_dif'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff()
        df_feature_set_2_calculated[f'{variable.lower()}_mean_dif'] = feature_data_raw.groupby(['fyearq', 'fqtr'])[f'{variable.lower()}_dif'].mean()

    if 'vwmean_yoy' in calculations:
        # Value-weighted average der Year-over-Year growth rate ("vwmean_yoy")
        if f'{variable.lower()}_yoy' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_yoy'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff(4)
        feature_data_raw[f'{variable.lower()}_yoy_mkvaltq'] = (feature_data_raw[f'{variable.lower()}_yoy'] * feature_data_raw['mkvaltq'])
        df_feature_set_2_calculated[f'{variable.lower()}_vwmean_yoy'] = feature_data_raw.groupby(['fyearq', 'fqtr'])[f'{variable.lower()}_yoy_mkvaltq'].sum()/feature_data_raw.groupby(['fyearq', 'fqtr'])['mkvaltq'].sum()

    if 'mean_yoy' in calculations:
        # Cross-sectional equal-weighted average der Year-over-Year growth rate ("mean_yoy")
        if f'{variable.lower()}_yoy' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_yoy'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff(4)
        df_feature_set_2_calculated[f'{variable.lower()}_mean_yoy'] = feature_data_raw.groupby(['fyearq', 'fqtr'])[f'{variable.lower()}_yoy'].mean()

# Zeige das aktualisierte DataFrame für Feature Set 3 an
df_feature_set_2_calculated.to_csv("data/feature_sets/feature_set_2_calculated.csv")
df_feature_set_2_calculated
```

```python
# Compute Variants for feature set 3
# DataFrame für die neuen Werte
df_feature_set_3_calculated = pd.DataFrame()

# Berechnung der spezifizierten Wachstumsmaße für jede Variable
for variable, calculations in tqdm(berechnungen_feature_set3.items()):
    if 'sd_dif' in calculations:
        # Standardabweichung der Quarter-over-Quarter growth rate ("sd_dif")
        if f'{variable.lower()}_dif' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_dif'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff()
        if f'{variable.lower()}_dif_atq' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_dif_atq'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff() / feature_data_raw['atq']
        if f'{variable.lower()}_dif_saleq' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_dif_saleq'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff() / feature_data_raw['saleq']

        df_feature_set_3_calculated[f'{variable.lower()}_sd_dif'] = feature_data_raw.groupby(['fyearq', 'fqtr'])[f'{variable.lower()}_dif'].std()


    if 'sd_yoy' in calculations:
        # Standardabweichung der Year-over-Year growth rate ("sd_yoy")
        if f'{variable.lower()}_yoy' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_yoy'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff(4)
        if f'{variable.lower()}_yoy_atq' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_yoy_atq'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff(4) / feature_data_raw['atq']
        if f'{variable.lower()}_yoy_saleq' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_yoy_saleq'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff(4) / feature_data_raw['saleq']

        df_feature_set_3_calculated[f'{variable.lower()}_sd_yoy'] = feature_data_raw.groupby(['fyearq', 'fqtr'])[f'{variable.lower()}_yoy'].std()


    if 'vwmean_dif' in calculations:
        # Value-weighted average der Quarter-over-Quarter growth rate ("vwmean_dif")
        if f'{variable.lower()}_dif' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_dif'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff()
        if f'{variable.lower()}_dif_atq' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_dif_atq'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff() / feature_data_raw['atq']
        if f'{variable.lower()}_dif_saleq' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_dif_saleq'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff() / feature_data_raw['saleq']

        feature_data_raw[f'{variable.lower()}_dif_mkvaltq'] = (feature_data_raw[f'{variable.lower()}_dif'] * feature_data_raw['mkvaltq'])
        df_feature_set_3_calculated[f'{variable.lower()}_vwmean_dif'] = feature_data_raw.groupby(['fyearq', 'fqtr'])[f'{variable.lower()}_dif_mkvaltq'].sum()/feature_data_raw.groupby(['fyearq', 'fqtr'])['mkvaltq'].sum()


    if 'mean_dif' in calculations:
        # Cross-sectional equal-weighted average der Quarter-over-Quarter growth rate ("mean_dif")
        if f'{variable.lower()}_dif' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_dif'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff()
        if f'{variable.lower()}_dif_atq' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_dif_atq'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff() / feature_data_raw['atq']
        if f'{variable.lower()}_dif_saleq' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_dif_saleq'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff() / feature_data_raw['saleq']

        df_feature_set_3_calculated[f'{variable.lower()}_mean_dif'] = feature_data_raw.groupby(['fyearq', 'fqtr'])[f'{variable.lower()}_dif'].mean()


    if 'vwmean_yoy' in calculations:
        # Value-weighted average der Year-over-Year growth rate ("vwmean_yoy")
        if f'{variable.lower()}_yoy' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_yoy'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff(4)
        if f'{variable.lower()}_yoy_atq' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_yoy_atq'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff(4) / feature_data_raw['atq']
        if f'{variable.lower()}_yoy_saleq' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_yoy_saleq'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff(4) / feature_data_raw['saleq']

        feature_data_raw[f'{variable.lower()}_yoy_mkvaltq'] = (feature_data_raw[f'{variable.lower()}_yoy'] * feature_data_raw['mkvaltq'])
        df_feature_set_3_calculated[f'{variable.lower()}_vwmean_yoy'] = feature_data_raw.groupby(['fyearq', 'fqtr'])[f'{variable.lower()}_yoy_mkvaltq'].sum()/feature_data_raw.groupby(['fyearq', 'fqtr'])['mkvaltq'].sum()


    if 'mean_yoy' in calculations:
        # Cross-sectional equal-weighted average der Year-over-Year growth rate ("mean_yoy")
        if f'{variable.lower()}_yoy' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_yoy'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff(4)
        if f'{variable.lower()}_yoy_atq' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_yoy_atq'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff(4) / feature_data_raw['atq']
        if f'{variable.lower()}_yoy_saleq' not in feature_data_raw.columns:
            feature_data_raw[f'{variable.lower()}_yoy_saleq'] = feature_data_raw.groupby('gvkey')[f'{variable.lower()}'].diff(4) / feature_data_raw['saleq']

        df_feature_set_3_calculated[f'{variable.lower()}_mean_yoy'] = feature_data_raw.groupby(['fyearq', 'fqtr'])[f'{variable.lower()}_yoy'].mean()


# Zeige das aktualisierte DataFrame für Feature Set 3 an
df_feature_set_3_calculated.to_csv("data/feature_sets/feature_set_3_calculated.csv")
df_feature_set_3_calculated
```
