```python
import pandas as pd
```

```python
lag_target_df = pd.read_excel('data/lag_and_target.xlsx', sheet_name='DATA')
lag_target_df
```

```python
NGDP_mean_df = pd.read_excel('data/MeanMedian_NGDP_Growth.xlsx', sheet_name='Mean_Growth')
NGDP_mean_df.rename(columns={"DNGDP2": "DNGDP2_mean", "DNGDP3": "DNGDP3_mean","DNGDP4": "DNGDP4_mean","DNGDP5": "DNGDP5_mean","DNGDP6": "DNGDP6_mean"}, inplace=True)
NGDP_mean_df
```

```python
NGDP_median_df = pd.read_excel('data/MeanMedian_NGDP_Growth.xlsx', sheet_name='Median_Growth')
NGDP_median_df.rename(columns={"DNGDP2": "DNGDP2_median", "DNGDP3": "DNGDP3_median","DNGDP4": "DNGDP4_median","DNGDP5": "DNGDP5_median","DNGDP6": "DNGDP6_median"}, inplace=True)
NGDP_median_df
```

```python
RGDP_mean_df = pd.read_excel('data/MeanMedian_RGDP_Growth.xlsx', sheet_name='Mean_Growth')
RGDP_mean_df.rename(columns={"RGDP2": "RGDP2_mean", "RGDP3": "RGDP3_mean","RGDP4": "RGDP4_mean","RGDP5": "RGDP5_mean","RGDP6": "RGDP6_mean"}, inplace=True)
RGDP_mean_df
```

```python
RGDP_median_df = pd.read_excel('data/MeanMedian_RGDP_Growth.xlsx', sheet_name='Median_Growth')
RGDP_median_df.rename(columns={"RGDP2": "RGDP2_median", "RGDP3": "RGDP3_median","RGDP4": "RGDP4_median","RGDP5": "RGDP5_median","RGDP6": "RGDP6_median"}, inplace=True)
RGDP_median_df
```

```python
frames = [NGDP_mean_df,
          NGDP_median_df[['DNGDP2_median', 'DNGDP3_median', 'DNGDP4_median', 'DNGDP5_median', 'DNGDP6_median']],
          RGDP_mean_df[['RGDP2_mean', 'RGDP3_mean', 'RGDP4_mean', 'RGDP5_mean', 'RGDP6_mean']],
          RGDP_median_df[['RGDP2_median', 'RGDP3_median', 'RGDP4_median', 'RGDP5_median', 'RGDP6_median']]
          ]
feature_set_1_without_lag = pd.concat(frames, axis=1, )

feature_set_1_without_lag = feature_set_1_without_lag[feature_set_1_without_lag['YEAR'] >= 1989]
feature_set_1_without_lag = feature_set_1_without_lag[feature_set_1_without_lag['YEAR'] <= 2022]

feature_set_1_without_lag.to_csv('data/feature_sets/feature_set_1_without_lag.csv')
feature_set_1_without_lag
```

```python
feature_set_1_lag = lag_target_df[lag_target_df.columns[:4]]
feature_set_1_lag = feature_set_1_lag.shift(1)

feature_set_1_lag = feature_set_1_lag[feature_set_1_lag['Year'] >= 1989]
feature_set_1_lag = feature_set_1_lag[feature_set_1_lag['Year'] <= 2022]

feature_set_1_lag.to_csv('data/feature_sets/feature_set_1_lag.csv')
feature_set_1_lag
```

```python
target = lag_target_df[['Year', 'Quarter', 'Third']]

target = target[target['Year'] >= 1989]
target = target[target['Year'] <= 2022]

target.to_csv('data/target.csv')
target
```
